```python
import time

def work(n):
    time.sleep(1)
    return n * n

results = []
for i in range(10):
    results.append(work(i))

print(results)
```

```python
import asyncio

async def work(n):
    await asyncio.sleep(1)
    return n * n

async def main():
    tasks = [work(i) for i in range(10)]
    results = await asyncio.gather(*tasks)
    print(results)

await main()
```
